```python
import pandas as pd
import requests
from python.nhtsa_vin_decoder import NHTSAVinDecoder     # Library to decode the vins and get their engine size 
```

```python
# Read the csv file with all h numbers and engine sizes
df_vins = pd.read_csv("C:\Language_Projects\Language_Projects\Python\Flagship_1\\automotive_diagnostic_data_cleaning\data\VIN Lookup.csv")
```

```python
# Print the head to get one example
df_vins_head = df_vins.head(5)
```

```python
# Decode one VIN# as example
decoder = NHTSAVinDecoder()
vehicle = decoder.decode('1J8GP28K08W103545')
```

```python
# Print the engine size found
vehicle.engine_displacement_l
```

```python
# List to save all engine sizes
engine_size_list = []

# Loop to iterate over the df vin column
for vin in df_vins['Customer VIN']:
    # Decode the engine sizes and save them into the engine size list
    decoder = NHTSAVinDecoder()
    vehicle = decoder.decode(vin)
    # Append the engine sizes during iteration and add a L to create the main df engine size format
    engine_size_list.append(f"{vehicle.engine_displacement_l}L")
```

```python
# Create a new column to update the df with the engine sizes
df_vins['engine_size'] = engine_size_list
```

```python
# Export to .csv to update to the main the df 
df_vins.to_csv("C:\Language_Projects\Language_Projects\Python\Flagship_1\\automotive_diagnostic_data_cleaning\data\df_engine_sizes.csv", index=False)
```
